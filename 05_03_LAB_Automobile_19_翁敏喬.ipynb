# Number vs Number

```python
# 找出引擎大小(Engine-size)跟馬力(horsepower)之間有沒有關係？

# 方法一使用散佈圖（Scatter Plot）觀察關係
# 解釋：
# 散佈圖（Scatter Plot）可以幫助我們觀察兩者之間是否有線性關係
# 如果點的分佈呈現上升趨勢，表示引擎越大，馬力越高

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Engine-size", y="horsepower", data=data)
plt.title("Engine Size vs Horsepower")
plt.xlabel("Engine Size")
plt.ylabel("Horsepower")
plt.show()

# 方法二使用回歸圖（Regression Plot）看線性趨勢
# 解釋：
# regplot會加上回歸線，幫助我們觀察數據是否呈現線性關係

plt.figure(figsize=(8, 6))
sns.regplot(x="Engine-size", y="horsepower", data=data)
plt.title("Regression: Engine Size vs Horsepower")
plt.xlabel("Engine Size")
plt.ylabel("Horsepower")
plt.show()
```

```python
# 計算引擎大小(Engine-size)跟馬力(horsepower)之間相關性係數

correlation = data["Engine-size"].corr(data["horsepower"])
print(f"Engine Size 與 Horsepower 的相關係數：{correlation:.2f}")
```

解釋：

相關係數 r 值範圍：

r > 0.7：高度正相關（引擎越大，馬力越高）

0.3 < r < 0.7：中度相關

r < 0.3：低度相關

r ≈ 0：幾乎無關

```python
# 練習題
# 找出價格跟馬力之間有沒有關係？

# 檢查資料中是否有缺失值
data = data.dropna(subset=['price', 'horsepower'])

# 繪製散點圖
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='price', data=data)
plt.title('Price vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.show()

# 計算價格與馬力之間的皮爾森相關係數
correlation = data['horsepower'].astype(float).corr(data['price'])
print(f'Price and Horsepower Pearson correlation: {correlation}')
```

這段代碼首先清理了 price 和 horsepower 兩個欄位中的缺失值，

接著繪製了散點圖，並計算了這兩者之間的皮爾森相關係數。

如果散點圖顯示出明顯的線性關係（例如，大多數點沿著一條線排列），

那麼價格和馬力之間可能存在較強的關聯。
如果皮爾森相關係數接近 1 或 -1，

表示它們之間有較強的正或負線性關聯；

如果接近 0，則表明關聯較弱。

```python
# 練習題
# 找出價格跟車長之間有沒有關係？

# 檢查資料中是否有缺失值
data = data.dropna(subset=['price', 'length'])

# 繪製散點圖
plt.figure(figsize=(10, 6))
sns.scatterplot(x='length', y='price', data=data)
plt.title('Price vs Length')
plt.xlabel('Length')
plt.ylabel('Price')
plt.show()

# 計算價格與車長之間的皮爾森相關係數
correlation = data['length'].astype(float).corr(data['price'])
print(f'Price and Length Pearson correlation: {correlation}')
```

1.處理缺失值：清理 price 和 length 欄位中的缺失值。

2.繪製散點圖：展示價格和車長之間的直觀關係。

3.計算皮爾森相關係數：得出兩者之間的數字化相關性，

範圍是 -1 到 1。

如果發現散點圖有明顯的線性趨勢或皮爾森相關係數接近 1 或 -1，

那麼就可以推測價格和車長之間存在一定的關係。

如果相關係數接近 0，則兩者之間的關係可能較弱。

# Label vs Label

```python
# 找出Body-style跟Num-of-doors的關係

# 檢查資料中是否有缺失值
data = data.dropna(subset=['Body-style', 'Num-of-doors'])

# 計算交叉表，顯示不同車身風格和門數的頻次
cross_tab = pd.crosstab(data['Body-style'], data['Num-of-doors'])
print(cross_tab)

# 繪製條形圖
plt.figure(figsize=(10, 6))
sns.countplot(x='Body-style', hue='Num-of-doors', data=data)
plt.title('Body-style vs Num-of-doors')
plt.xlabel('Body-style')
plt.ylabel('Count')
plt.show()
```

1.處理缺失值：清理 body-style 和 num-of-doors 欄位中的缺失值。

2.交叉表：用 pd.crosstab 計算 body-style 和 num-of-doors 之間的頻率分佈，

這樣可以看到每個車身風格對應的門數分佈。

3.條形圖：使用 countplot 畫出條形圖，

視覺化不同車身風格和門數的關係。

可以根據交叉表來查看不同車身風格（例如，轎車、SUV、跑車等）

與門數之間的具體關係。條形圖將顯示每種車身風格下，門數的分佈情況。

```python
# 計算Body-style跟Num-of-doors的計數矩陣

# 方法一
# 使用 pd.crosstab 生成 Body-style 和 Num-of-doors 的交叉表
cross_tab = pd.crosstab(data['Body-style'], data['Num-of-doors'])
# 顯示交叉表
print(cross_tab)

# 方法二
data.groupby(['Body-style','Num-of-doors']).size().unstack(fill_value=0)
```

我們可以使用pd.crosstab來生成交叉表，

這將顯示每一種車身風格（Body-style）和每一種門數（Num-of-doors）

的組合出現的次數。

這段代碼會生成一個計數矩陣，顯示每一種車身風格（例如轎車、SUV等）

對應每一種門數（例如兩門或四門）的數量。

矩陣的行表示不同的車身風格，列表示不同的門數。

每個單元格中的數字是該組合出現的次數。

```python
# 畫出上個表格的熱圖

# 方法一使用 seaborn 繪製熱圖
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, cmap="YlGnBu", fmt="d", linewidths=0.5)
plt.title('Heatmap of Body-style vs Num-of-doors')
plt.xlabel('Num-of-doors')
plt.ylabel('Body-style')
plt.show()

# 方法二
matrix = data.groupby(['Body-style','Num-of-doors']).size().unstack(fill_value=0)
sns.heatmap(matrix, annot=True, cmap='coolwarm', cbar=True)
plt.show()
```

方法一:

sns.heatmap()：生成熱圖。

annot=True：在每個單元格中顯示數字（即計數）。

cmap="YlGnBu"：設置熱圖的顏色方案，可以根據需要更改為其他顏色方案。

fmt="d"：確保顯示的數字是整數。

linewidths=0.5：設置每個單元格之間的邊界線寬度。

plt.title()、plt.xlabel() 和 plt.ylabel()：設置圖表的標題和軸標籤。

```python
# 練習題
# 找出Body-style跟廠牌的關係

# 計算 Body-style 和 make（廠牌）之間的交叉表
cross_tab = pd.crosstab(data['Body-style'], data['make'])

# 顯示交叉表
print(cross_tab)

# 繪製 Body-style 和 make 之間關係的熱圖
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, annot=True, cmap="YlGnBu", fmt="d", linewidths=0.5)
plt.title('Heatmap of Body-style vs Make')
plt.xlabel('Make')
plt.ylabel('Body-style')
plt.show()
```

# Group By

```python
# 計算不同廠牌汽車的平均價格，並畫出來

# 使用 groupby 計算不同廠牌汽車的平均價格
avg_price_by_make = data.groupby('make')['price'].mean()

# 顯示計算結果
print(avg_price_by_make)

# 繪製條形圖
plt.figure(figsize=(14, 8))
avg_price_by_make.plot(kind='bar', color='skyblue')
plt.title('Average Car Price by Make')
plt.xlabel('Make')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

1.groupby('make')['price'].mean()：將數據按 make（廠牌）分組，

並計算每個廠牌的平均價格。

2.avg_price_by_make.sort_values(ascending=False).plot(kind='bar', color='skyblue')：

將計算出來的平均價格按降序排列，並繪製條形圖。顏色設置為skyblue。

3.plt.xticks(rotation=90)：將 x 軸標籤旋轉 90 度，以便更清楚地顯示廠牌名稱。

```python
# 練習題
# 修改上圖將平均價格由高到低排序

# 使用 groupby 計算不同廠牌汽車的平均價格
avg_price_by_make = data.groupby('make')['price'].mean()

# 按照平均價格從高到低排序
avg_price_by_make_sorted = avg_price_by_make.sort_values(ascending=False)

# 顯示排序後的平均價格
print(avg_price_by_make_sorted)

# 繪製排序後的條形圖
plt.figure(figsize=(14, 8))
avg_price_by_make_sorted.plot(kind='bar', color='skyblue')
plt.title('Average Car Price by Make (Sorted from High to Low)')
plt.xlabel('Make')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

1.要將上圖的平均價格由高到低排序，你只需要對avg_price_by_make進行排序，

並在繪圖時使用排序後的數據。這樣條形圖將會按照價格從高到低顯示廠牌。

2.avg_price_by_make.sort_values(ascending=False)：

將平均價格按降序排序，確保條形圖從價格高的廠牌開始顯示。

3.繪製條形圖時，顯示的廠牌將按照價格從高到低排列。

```python
# 練習題
# 計算不同車型的平均馬力

# 使用groupby計算不同車型的平均馬力
avg_horsepower_by_body_style = data.groupby('Body-style')['horsepower'].mean()

# 顯示結果
print(avg_horsepower_by_body_style)

# 繪製不同車型的平均馬力條形圖
plt.figure(figsize=(10, 6))
avg_horsepower_by_body_style.sort_values(ascending=False).plot(kind='bar', color='lightcoral')
plt.title('Average Horsepower by Body-style')
plt.xlabel('Body-style')
plt.ylabel('Average Horsepower')
plt.xticks(rotation=45)
plt.show()
```
